### Problem solving

```python
def rev_number(n):
    s = 0
    while True:
        k = str(n)
        if k == k[::-1]:
            break
        else:
            m = int(k[::-1])
            n += m
            s += 1
            return n
```

```python
s = input()
list1 = s.split(",")
list2 = [i.split(":")[1] for i in list1]
list3 = []
for num in list2:#num ='750'
    n = int(num) #n=750
    rev= num[::-1]#rev=57
    while str(n) != rev:
        n += int(rev)
        rev = str(n)[::-1]
    list3.append(n)
list1[list3.index(max(list3))].split(":")[0]
```
