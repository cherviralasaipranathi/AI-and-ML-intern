```python
# Encode the categorical columns using different kinds of encoding techniques from pandas and scikit-learn
merge_d=pd.get_dummies(df,columns=['distance','consume','speed','temp_inside','temp_outside','specials','gas_type','AC','rain','sun','refill liters','refill gas'])
merge_d
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # creating an object
merge_d=sc.fit_transform(merge_d)
merge_d
```

```python
merge_d=pd.DataFrame(merge_d)
merge_d
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[('Encoder',OneHotEncoder(),['gas_type'])],remainder="passthrough")
df1=ct.fit_transform(df)
df1=pd.DataFrame(df1)
df1=df1.iloc[:,0:2]
df1.columns=['E98','SP10']
df1=pd.concat([df,df1],axis=1)
df1=df1.drop('gas_type',axis=1)
df1
```
