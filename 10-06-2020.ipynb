#### Relational 
- b/n two numerical columns
- relplot
- Scatterplot -relplot(kind=â€™scatterâ€™)
- lineplot

```python
sns.relplot("total_bill","tip",data=tips)
```

```python
sns.relplot("total_bill","tip",data=tips,hue='smoker')
```

```python
sns.relplot("total_bill","tip",data=tips,col='smoker')
```

```python
sns.relplot("total_bill","tip",data=tips,hue='smoker',col='sex')
```

```python
sns.relplot("total_bill","tip",data=tips,hue='smoker',style='smoker')
```

```python
sns.relplot("total_bill","tip",data=tips,kind='scatter')
```

```python
sns.relplot("total_bill","tip",data=tips,kind='line')
```
